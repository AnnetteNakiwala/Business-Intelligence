```python
# Movie analysis using python libraries to explore and analyze large dataset and giving an overview of movies

#importing libraries
import pandas as pd #creating dataframes
import numpy as np # big data list
import matplotlib.pyplot as plt # same as from matplotlib import pyplot as plt 

#load the data from movies.csv into a pandas dataframe using read_csv() function
df = pd.read_csv('movies.csv')

#Diplay data
df.head()
```

```python
# Understanding your data
df.info()
```

```python
# Data Cleaning and preprocessing: includes checking of missing values, removing duplicates, misplaced data

#check for missing values
df.isnull().sum()

```

```python
#Removing null values using a function called dropna()
df = df.dropna()
df.isnull().sum()

```

```python
# Remove duplicate values using drop_duplicate() function
df = df.drop_duplicates() # or : df.drop_duplicates(inplace = True)
```

```python
# Convert release data from datatype(Object) to date time
df['release_date'] = pd.to_datetime(df['release_date'])
df.info()
```

```python
# Stastical Analysis: Variance, STD, Mean, Min, Max
df.describe()
```

```python
#Explorational Data Analysis using pandas and matplotlib the explore the data ( Creating visualisation such as histograms)
#create a histogram of movie ratings
plt.hist(df['vote_count'], bins= 5)
plt.ylabel("Frequency")
plt.xlabel("Ratings")
plt.title("Distribution of Movie Ratings")
plt.plot()
```

```python
# Data Aggregation and Grouping using groupby() function

#group the data by genre and calculate the average ratings
genre_ratings = df.groupby('genres')['vote_count'].agg(['mean', 'std', 'count'])

#Data Visualization ang grouping
#create a bar plot
genre_ratings.plot(kind = 'bar', y ='mean', yerr = 'std' )
plt.show()
```

```python
# Machine Learning and predictive analysis using Linear regression
#import machine leanring library
from sklearn.linear_model import LinearRegression

x = df[['budget', 'runtime']]
y = df['revenue']

#build the model
model = LinearRegression()
x = x.values # converting the x into an array
model.fit(x,y)

#Predict revenue for a new moview
new_movie = [[10000000, 120]]
predicted_revenue = model.predict(new_movie)
print("Predicted Revenue for the movie is :",predicted_revenue)
```
